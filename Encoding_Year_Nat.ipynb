```python
import pickle as pkl
import pandas as pd
import numpy as np

import re
import matplotlib.pyplot as plt
import seaborn as sns

import datetime as dt
from datetime import datetime

plt.rcParams["figure.figsize"] = (10,8)
pd.options.display.max_rows=None
```

```python
df = pd.read_pickle("./Datos/bikes_eda_limpio.pkl")
df.sample(10)
```

```python
df_copia = df.copy()
```

```python
df_copia.year.unique()
```

```python
df_copia.info()
```

Sacamos el DataFrame con el año codificado para los clientes "registered"

```python
df_registered = df_copia.drop(['casual'], axis=1)
df_registered.head()
```

```python
lista = ["year"]

df_encoded = pd.DataFrame()


for columna in lista:
    df_dummies = pd.get_dummies(df_registered[columna], prefix_sep = "_", prefix = columna, dtype = int)

    df_encoded = pd.concat([df_encoded, df_dummies], axis = 1)
```

```python
df_encoded.head()
```

```python
df_final_registered = pd.concat([df_registered, df_encoded], axis = 1)
df_final_registered.head()
```

```python
df_final_registered.drop(lista, axis = 1, inplace=True)
df_final_registered.head(2)
```

Sacamos el DataFrame con el año codificado para los clientes "casual"

```python
df_casual = df_copia.drop(['registered'], axis=1)
df_casual.head()
```

```python
lista = ["year"]

df_encoded = pd.DataFrame()


for columna in lista:
    df_dummies = pd.get_dummies(df_casual[columna], prefix_sep = "_", prefix = columna, dtype = int)

    df_encoded = pd.concat([df_encoded, df_dummies], axis = 1)
```

```python
df_encoded.head()
```

```python
df_final_casual = pd.concat([df_casual, df_encoded], axis = 1)
df_final_casual.head()
```

```python
df_final_casual.drop(lista, axis = 1, inplace=True)
df_final_casual.head(2)
```
